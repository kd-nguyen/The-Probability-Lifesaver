```python
# My own solution to the birthday problem: how many people do we need to have
# in a room before there's at least a 50% chance that two share a birthday?
def same_birthday_prob(n, days):
    """Calculate the probability of 2 people having the same birthday from
    a group of n people"""
    prop = 0
    for n_i in range(1, n):
        nom = n_i
        for m in range(days, days - n_i, -1):
            nom *= m
        denom = days ** (n_i + 1)
        prop += nom / denom
    
    return prop

same_birthday_prob(3, 365)
```

```python
def no_same_birthday_prob(n, days):
    """Calculate the probability of none in a group of n shares the same BD"""
    nom = math.factorial(days)
    denom = days ** n * math.factorial(days - n)
    return nom / denom
```

```python
# Answer: 23
x = range(1, 50)
days = 687
y1 = list(map(lambda x: same_birthday_prob(x, days), x))
y2 = list(map(lambda x: no_same_birthday_prob(x, days), x))
plt.figure(figsize = (15, 5))
plt.plot(x, y1, label='At least 2 shares BD')
plt.plot(x, y2, label='None shares BD')
plt.axhline(0.5, c = 'red')
plt.xticks(x)
plt.legend()
plt.show()
```

```python
# Power series exploration (Larson Calculus p.659)
def f_gen(f_term, x, n):
    term_sum = 0
    for n_i in range(n + 1):
        term_sum += f_term(x, n_i)
    
    return term_sum
```

```python
def f1(x, n):
    return (-1)**n * x**n / math.factorial(n)
```

```python
def f2(x, n):
    return (-1)**n * x**(2*n) / math.factorial(2*n)
```

```python
def f3(x, n):
    return (-1)**n * x**(2*n + 1) / math.factorial(2*n + 1)
```

```python
def f4(x, n):
    return (-1)**n * x**(2*n + 1) / (2*n + 1)
```

```python
def f5(x, n):
    return 2**n * x**n / math.factorial(n)
```

```python
def draw(f, x_range, n):
    x = list(x_range)
    y = list(map(lambda x: f_gen(f, x, n), x))
    plt.plot(x, y, label = f.__name__)

x = np.arange(-2, 2, 0.1)
for func in [f1, f2, f3, f4, f5]:
    draw(func, x, 10)

plt.ylim(0, 3)
plt.legend()
plt.show()
```
